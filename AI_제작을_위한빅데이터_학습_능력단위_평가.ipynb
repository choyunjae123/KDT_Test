```python
df_seoul_gangnam = df_seoul[df_seoul['시군구명'] == '강남구']
df_seoul_gangnam
```

```python
df_mcdonald_gangnam = df_seoul_gangnam[df_seoul_gangnam['상호명'].str.contains(r'맥도날드')]
df_mcdonald_gangnam
```

```python
# '강남구'에 '맥도날드'를 지도에 표기
# 2점
!pip install folium
import folium
```

```python
map = folium.Map(location=[37.50006, 127.0355806], zoom_start=15)
for _, row in df_mcdonald_gangnam.iterrows():
  popup = folium.Popup(row['상호명'], max_width=200)
  folium.Marker(location=[row['위도'], row['경도']], popup=row['상호명']).add_to(map)
map
```

```python
# '강남구'에 패스트푸드 매장을 지도에 표기
# 2점
df_fastfood_gangnam = df_seoul_gangnam[df_seoul_gangnam['상호명'].str.contains(r'롯데리아|맥도날드|버거킹|맘스터치')]
df_fastfood_gangnam
```
